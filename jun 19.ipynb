```python
df = pd.DataFrame({
    "Age":[20,25,30,35,40],
    "Salary":[30000,40000,50000,60000,70000],
    "Experience":[1,2,3,4,5],
    "Country":["India","India","India","India","India"],
    "Purchased":[0,0,1,1,1]
})
```

```python
df
```

```python
#using the corr method findding which features are best to select based on corr
df.corr(numeric_only=1)
```

```python
#using the selectKbest method
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
```

```python
selector = SelectKBest(score_func=f_classif, k=2)
```

```python
X = df.drop(["Purchased", "Country"], axis=1)
```

```python
y = df['Purchased']
```

```python
selector.fit_transform(X, y)
```

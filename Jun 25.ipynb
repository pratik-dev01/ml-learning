```python
for i in cols: 
    Q1 = X_train[i].quantile(0.25)
    Q3 = X_train[i].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    X_train = X_train[(X_train[i] > lower) & (X_train[i] < upper)]
```

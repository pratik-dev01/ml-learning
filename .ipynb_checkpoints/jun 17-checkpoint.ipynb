```python
df = pd.DataFrame({
    "City": ["Pune", "Mumbai", None, "Pune", "Delhi"]
})
```

```python
mod = df['City'].mode()[0]
```

```python
df = pd.read_csv("outliers.csv")
```

```python
Q1 = df['Age'].quantile(0.25)
```

```python
Q3 = df['Age'].quantile(0.75)
```

```python
IQR = Q3 - Q1
```

```python
print(Q1, Q3, IQR)
```

```python
lower = Q1 - 1.5*IQR
```

```python
upper = Q3 + 1.5*IQR
```

```python
print(lower, upper)
```

```python
print((120 < lower or 120 > upper))
```

```python
Outliers = df[(df['Age'] < lower) | (df['Age'] > upper)]
```

```python
Outliers
```
